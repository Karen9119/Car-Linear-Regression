## Data Preparation

```python
df.head()
```

```python
# extract all columns with the data types as object
cat = df.select_dtypes(include = ['object'])
cat.head()
```

## Creating Dummy Variables

```python
# create dummy variables for all these columns
cat_dummies = pd.get_dummies(cat, drop_first = True)
```

```python
# let's drop all the categorical columns from the df dataframe.
df.drop(columns = list(cat.columns), inplace = True)
```

```python
df = pd.concat([df, cat_dummies], axis = 1)
```

```python
df.head()
```

## Split the data into Train-Test

```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df, train_size = 0.7, random_state = 100)
```

```python
df_train.head()
```

```python
df_test.head()
```

## Scaling

```python
num = df.select_dtypes(include = ['int64', 'float64'])
num.head()
```

```python
num_var = list(num.columns)
```

```python
print(num_var)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df_train[num_var] = scaler.fit_transform(df_train[num_var])
```

```python
df_train.head()
```

#### Split into X and y

```python
y_train = df_train.pop('price')
x_train = df_train
```

## ML Modelling

### Modeling based on RFE 

```python
from sklearn.feature_selection import RFE

lm = LinearRegression()

rfe = RFE(lm, 12)

rfe.fit(x_train, y_train)
```

```python
x_train.columns
```

```python
rfe.support_
```

```python
col = list(x_train.columns[rfe.support_])
col
```

```python
x_train_rfe = x_train[col]
x_train_rfe.head()
```

### Manual Modelling: StatsModel

```python
import statsmodels.api as sm
x_train_rfe = sm.add_constant(x_train_rfe)

lm1 = sm.OLS(y_train, x_train_rfe).fit()
print(lm1.summary())
```

**Price**

- Brand Of the car
- Insurance Rating
- Engine Size

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
x_train_rfe.head()
```

```python
x_train_rfe.drop(columns = 'const').values
```

```python
x_train_vif = x_train_rfe.drop(columns = 'const', axis = 1)
vif = pd.DataFrame()
vif['features'] = x_train_vif.columns
vif['VIF'] = [variance_inflation_factor(x_train_vif.values, i) for i in range(len(x_train_vif.columns))]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by ='VIF', ascending =False)
vif
```

## Model Evaluation

```python
df_test.head()
```

```python
df_test[num_var] = scaler.transform(df_test[num_var])
```

```python
df_test.head()
```

```python
y_test = df_test.pop('price')
x_test = df_test
```

```python
retain = list(x_train_rfe.columns[1:])
```

```python
print(retain)
```

```python
x_test_rfe = x_test[retain]
x_test_rfe.head()
```

```python
x_test_rfe = sm.add_constant(x_test_rfe)
```

```python
y_pred = lm1.predict(x_test_rfe)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test,y_pred)
```

```python
sns.regplot(x=y_test, y=y_pred,data=None, color='blue').set(xlabel='Test price', ylabel='Predicted Price', title= 'y_test VS y_pred')
```
